```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.inspection import permutation_importance
from sklearn.metrics import confusion_matrix
#for multiple feature Gaussian is better to use
from sklearn.naive_bayes import GaussianNB
import pandas as pd
```

```python
data = pd.read_csv('breast-cancer-wisconsin.csv', na_values = '?')
data1 = data.dropna()
data1
```

```python
X = data1.iloc[:,1:10]
y = data1.iloc[:,10]
X_test, X_train,Y_test,Y_train = train_test_split(X, y, test_size = 0.3)
```

```python
#created classifier
GNB = GaussianNB()
#fit the model using train datasets
GNB.fit(X_train,Y_train)
```

```python
# to predict the class for test data set
y_pred = GNB.predict(X_test)
```

```python
y_test = Y_test.tolist()
x_test = pd.DataFrame(X_test)
x_test['Class'] = y_test
x_test['Predicted_class'] = y_pred
print(x_test.head(20))
```

```python
target = ['Benign', 'Malignant']
print('Accuracy Score is: ',accuracy_score(y_test, y_pred)) #TP+FN/TP+TF+FP+FN
print('\n\n\nclassification_report of data after Knn applied: \n',classification_report(y_test, y_pred,target_names=target),'\n\n') 
#precision = TP/TP+FP & recall = TP/TP+TN & f1_score = harmonic_mean of pre and recall 
print('Confusion Matrix of data predicted \n',confusion_matrix(y_test, y_pred),'\n\n') #TP,TN,FP,FN,
print('The model with benign class has good precision(exactness) and recall(completeness) value along with high f1-score.')
```
